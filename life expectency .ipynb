## LIFE EXPECTENCY 

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url='https://www.worldometers.info/demographics/life-expectancy/'
```

```python
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.text)
soup.prettify
```

```python
table=soup.find('table',id="example2")
table
```

```python
table=table.find('tbody')
table
```

```python
rows=table.find_all('tr')
rows
```

```python
rows=table.find_all('tr',style="")
rows
```

```python
column_info=[]
for i in rows:
    cols=i.find_all('td')
    #print(cols)
    country_info=[c.text for c in cols]
    print(country_info)
    column_info.append(country_info)
print(column_info)
```

```python
Number=[]
country=[]
Life_Expectency_both_sexes=[]
Females_lifeexpectency=[]
Males_lifeexpectency=[]
for i in column_info:
    Number.append(i[0])
    country.append(i[1])
    Life_Expectency_both_sexes.append(i[2])
    Females_lifeexpectency.append(i[3])
    Males_lifeexpectency.append(i[4])
```

```python
Data={'Number':Number,
     'country':country,
     'Life_Expectency_both_sexes':Life_Expectency_both_sexes,
     'Females_lifeexpectency':Females_lifeexpectency,
     'Males_lifeexpectency':Males_lifeexpectency}
```

```python
Life_expectency_info=pd.DataFrame(Data)
Life_expectency_info
```

```python
Life_expectency_info. to_csv('life_Expectency.csv',index=False)
```
